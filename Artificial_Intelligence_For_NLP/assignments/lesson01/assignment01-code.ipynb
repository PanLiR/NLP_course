```python
def create_grammar(grammar_str, split='=>', line_split='\n'):
    grammar = {}
    for line in grammar_str.split(line_split):
        if not line.strip(): continue
        exp, stmt = line.split(split)
        grammar[exp.strip()] = [s.split() for s in stmt.split('|')]
    return grammar
```

```python
import re
def token(string):
    return re.findall(r"[\u4e00-\u9fa5]", string)
```

```python
import jieba
def cut(string):
    return list(jieba.cut(string))
```

```python
def prob_1(word):
    if word in TOKEN:
        return words_count[word] / len(TOKEN)
    else:
        return 1/len(TOKEN)
```

```python
prob_1("我们")
```

```python
TOKEN = [str(t) for t in TOKEN]
```

```python
TOKEN
```

```python
TOKEN_2_GRAM = ["".join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-1]))]
```

```python
words_count_2 = Counter(TOKEN_2_GRAM)
```

```python
def prob_2(word1, word2):
    if word1 + word2 in words_count_2:
        return (words_count_2[word1+word2] / len(TOKEN_2_GRAM)) / prob_1(word1)
    else:
        return (1 / len(words_count_2)) / prob_1(word1)
```

```python
prob_2("我们", "在")
```

```python
def get_probability(sentence):

    sentence_pro = prob_1(sentence[0])
    
    for i, word in enumerate(sentence[:-1]):
        next_ = sentence[i+1]
        probability = prob_2(word, next_)
        sentence_pro *= probability
    
    return sentence_pro
```

```python
get_probability("今天吃了一碗拉面")
```

```python
get_probability("我要买人寿保险")
```

```python
import random
```

```python
def generate(gram, target):
    if target not in gram:
        return target
    expand = [generate(gram, t) for t in random.choice(gram[target])]
    return "".join(e if e != "\n" else "/n" for e in expand)
```

```python
def generate_n_probability(gram, target, n):
    res = []
    for i in range(10):
        sentence = generate(gram, target)
        probabilty = get_probability(sentence)
        res.append(tuple([sentence,probabilty]))
    return res
```
